```python
#importamos librerias 
import pandas as pd 
import numpy as np 
from pathlib import Path 
```

```python
#Links a los conjuntos de datos
url_cgsse = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/cgsse.CSV'
url_ocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV'
url_ft = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Fuerza%20de%20trabajo.CSV'
url_desocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV'

# CGSSE
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', 
                       low_memory=False)

#Ocupados
df_ocupados=pd.read_csv(url_ocupados,encoding='latin-1',sep=';',
                        low_memory=False)

#Fuerza de trabajo
df_ft=pd.read_csv(url_ft,
                  encoding='latin-1',sep=';',low_memory=False)
#No ocupados
df_desocupados=pd.read_csv(url_desocupados,encoding='latin-1',
                           sep=';',low_memory=False)
```

```python
#1. Población total
PT =  df_cgsse['FEX_C18'].sum()
print(f' R/ La población total es de {PT} millones de personas')

#Población en edad de trabajar
PET = df_cgsse[df_cgsse['P6040']>=15]['FEX_C18'].sum()
print(f' R/ La población en edad de trabajar es de {PET} millones de personas')

#Población que no está en edad de trabajar
PNET = df_cgsse[df_cgsse['P6040']<=15]['FEX_C18'].sum()
print(f' R/ La población que no está en edad de trabajar es de {PNET} millones de personas')

#Población económicamente activa
PEA=df_ft[(df_ft['FT']==1)]['FEX_C18'].sum()
print(f' R/ La población economicamente activa es de {PEA} millones de personas')

#FFT Pertenece a la PEI  si:1
PEI=df_ft[(df_ft['FFT']==1)]['FEX_C18'].sum()
print(f' R/ La población economicamente inactiva es de {PEI} millones de personas')

#OCI : Ocupado ? 1:si
O=df_ocupados[(df_ocupados['OCI']==1)]['FEX_C18'].sum()
TO=O/PET*100
print(f' R/ La tasa de ocupacion para diciembre es de {round(TO,1)}%')

#OCI : Ocupado ? 1:si
D=df_desocupados[(df_desocupados['DSI']==1)]['FEX_C18'].sum()
TD=D/PEA*100
print(f' R/ La tasa de desempleo para diciembre es de {round(TD,1)}%')
```
